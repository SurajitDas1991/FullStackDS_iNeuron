Write a Python program to do arithmetical operations addition and division.

```python
def MathProgram(a:int, b:int,operation:str):
    if operation == "Addition":
        return a+b
    elif  operation == "Division":
        return a/b
    else : 
        return "Operation not supported"

print(MathProgram(3,4,"Division"))
```

Write a Python program to find the area of a triangle

```python
def AreaOfATriangle(base:float,height:float):
    return 0.5*base*height
area=AreaOfATriangle(3,5)
print(f'Area of a Triangle is {area}')
```

Write a Python program to swap two variables

```python
x = 5
y = 10

print(f"Values before swapping FirstValue - {x} and SecondValue - {y}")
x, y = y, x
print(f"Values after swapping FirstValue - {x} and SecondValue - {y}")
```
