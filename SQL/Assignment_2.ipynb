Write an SQL query to find the position of the alphabet (‘a’) in the first name
column ‘Amitabh’ from Worker table.

```python
Select INSTR(FIRST_NAME, 'a') from Worker where FIRST_NAME = 'Amitabh';
```

Write an SQL query to print the name of employees having the highest salary in each
department.

```python
SELECT Worker.DEPARTMENT ,Worker.FIRST_NAME,Worker.SALARY FROM ( SELECT MAX(Salary) AS TOTAL_SALARY, Department
FROM Worker
GROUP BY Department) AS TEMPDB INNER JOIN Worker ON TEMPDB.DEPARTMENT=WORKER.DEPARTMENT and TEMPDB.TOTAL_SALARY=WORKER.Salary;
```
